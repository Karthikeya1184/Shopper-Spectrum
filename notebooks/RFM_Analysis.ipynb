# RFM Analysis

### Business Insights

**Recency**
- Customers with lower Recency values purchased more recently.
- These customers are more engaged and likely to respond to promotions.

**Frequency**
- Customers with high purchase frequency demonstrate loyalty.
- Retaining these customers should be a business priority.

**Monetary**
- High Monetary customers contribute a significant portion of total revenue.
- Premium offers and personalized marketing campaigns should target these customers.

### Why RFM Matters

RFM combines customer purchase behavior into three measurable metrics. These features will be standardized and used as inputs for the KMeans clustering algorithm to create actionable customer segments.
